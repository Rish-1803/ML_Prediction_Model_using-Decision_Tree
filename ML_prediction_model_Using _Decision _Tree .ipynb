###  *ML_ Prediction Model Using Decision Tree Method
### *  Predicting the salary of employees if more than 100k or not


```python
#Importing the relevant libraries and CSV file 
import pandas as pd
df = pd.read_csv('C:\\Users\\user\\Documents\\PythonPractiseFiles\\salaries.csv')
df
```

```python
#creating a variable named 'input' which includes whoile df and droping one column from it 'salary_more_then_100k'
#creating other variable with name 'target' which has column name 'salary_more_then_100k' (its dependent variable)

input = df.drop('salary_more_then_100k',axis= 'columns') 
target= df['salary_more_then_100k']
```

```python
#importing label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
#encoding  each column to make then numeric
le_company = LabelEncoder()
le_job   = LabelEncoder()
le_degree = LabelEncoder()
```

```python
#creating a new column with each variable as same name but label encoded

input['company_n'] = le_company.fit_transform(input['company'])
input['job_n']     = le_company.fit_transform(input['job'])
input['degree_n'] = le_company.fit_transform(input['degree'])
input
```

```python
#now using only label encoded colum and droping the columns with categorical value

input_n= input.drop(['company','job','degree'],axis ='columns')
input_n
```

```python
#from sklearn using tree  module to make decision on basis of different columns

from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
#fitting the model with both variables 'input' , and 'target'

model.fit(input_n,target)
```

```python
#checking thed score of the model
model.score(input_n,target)
```

```python
#getting prediction for abc pharma (0), sales executive(2),masters(1)
model.predict([[0,2,1]])
```

```python
#predicting thge value of google sales and master
model.predict([[2,2,1]])
```
